```python
import pandas as pd
import numpy as np

df = pd.read_csv("email_spam.csv", encoding='iso-8859-1')
```

```python
df.head()
```

```python
df.drop(labels=['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],inplace=True,axis=1)
df['v1'] = df['v1'].map({'ham':1,'spam':0})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split 
data_train, data_test = train_test_split(df,random_state=42,test_size=0.25)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import re
nltk.download('wordnet')
```

```python
corpus = []
for i in range (len(data_train['v2'])):
    document = re.sub(r'[^A-Za-z0-9]', ' ', data_train.iloc[i,1])
    document = [lemmatizer.lemmatize(word.lower(),pos='v') for word in document.split() if word not in stop_words]
    corpus.append(document)
```

```python
import gensim.downloader as api
model = api.load('word2vec-google-news-300')
```

```python
def avg_word2vec(corpus):
    output = []
    for document in corpus :
        vec = np.zeros((300,))
        for word in document :
            try: 
                vec = vec + model[word]
            except Exception as e:
                pass
        if (len(document) != 0) :
            vec = vec / len(document)
        else:
            vec = np.zeros((300,))
        output.append(vec)
    out = pd.DataFrame(output)
    return out

train_corpus = avg_word2vec(corpus)

```

```python
train_corpus
```

```python
train_target = data_train['v1'].values
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier(max_depth = 20, n_jobs = -1)
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
train_corpus = pca.fit_transform(train_corpus)

```

```python
print(train_corpus)
```

```python
RFC.fit(train_corpus,train_target)
```

```python
corpus = []
for i in range (len(data_test['v2'])):
    document = re.sub(r'[^A-Za-z0-9]', ' ', data_test.iloc[i,1])
    document = [lemmatizer.lemmatize(word.lower(),pos='v') for word in document.split() if word not in stop_words]
    corpus.append(document)
```

```python
test_corpus = avg_word2vec(corpus)
```

```python
test_target = data_test['v1'].values
```

```python
test_corpus
```

```python
test_corpus = pca.transform(test_corpus)
```

```python
pred = RFC.predict(test_corpus)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(test_target,pred)
print(classification_report(test_target,pred))
```
